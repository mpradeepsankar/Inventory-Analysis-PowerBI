```python
df['EmployeeHireDate'] = pd.to_datetime(df['EmployeeHireDate'])
df['OrderDate'] = pd.to_datetime(df['OrderDate'])

```

```python
df.dtypes
```

```python
df['EmployeePhone'] = df['EmployeePhone'].astype(str)
df['CustomerPhone'] = df['CustomerPhone'].astype(str)
```

```python
df.duplicated().sum()
```

```python
df['CategoryName'] = df['CategoryName'].str.strip().str.title()
df['Status'] = df['Status'].str.strip().str.title()
```

```python
df['Status']
```

```python
df.columns=df.columns.str.lower()

```

```python
df.columns
```

```python
df = df.rename(columns={
    'postalcode': 'postal_code',
    'employeeemail': 'employee_email',
    'employeephone': 'employee_phone',
    'productdescription': 'product_description',
    'customeraddress': 'customer_address',
    'customeremail': 'customer_email',
    'customerphone': 'customer_phone'
})
```

```python
df.columns
```

```python
def stock_status(qty):
    if qty <= 20:
        return 'Low Stock'
    elif qty <= 100:
        return 'Normal'
    else:
        return 'Overstock'

df['stock_status'] = df['total_item_quantity'].apply(stock_status)
```

```python
df['stock_status'].value_counts()
```

```python
df['stock_status'].value_counts().plot(kind='bar', title='Inventory Stock Status')
```

```python
overstocked_products = (
    df[df['stock_status'] == 'Overstock']
      .sort_values(by='total_item_quantity', ascending=False)
      .head(10)
)

overstocked_products[['product_name', 'total_item_quantity']]
```
